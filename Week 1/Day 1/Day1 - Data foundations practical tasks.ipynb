### 5. We can notice that for some features raw values differ significantly. Scale the data (except quality column) using Min-Max Scaling manually (create formula yourself). Then use built-in Min-Max Scaling function to validate results (display the resulting data frames separately).

Formula
$\Large x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}$

```python
features = wine_data.drop(columns=['quality'])
quality = wine_data['quality']

manual_scaled = (features - features.min())/(features.max() - features.min())
manual_scaled
```

### 11. Plot scatter plot for `quality` and column with strongest correlation to it. Does alcohol influence the quality of the wine?

```python

quality_corr = corr['quality'].drop('quality').abs().sort_values(ascending=False)
strongest_feature = quality_corr.index[0]

print(f"Strongest correlation to quality: {strongest_feature}({corr.loc[strongest_feature, 'quality']:.4f})")

plt.scatter(wine_data[strongest_feature], wine_data['quality'])
plt.xlabel(strongest_feature)
plt.ylabel('Quality')
plt.title(f'Quality vs {strongest_feature}')
plt.show()
```

**Interpretation**: Based on the correlation analysis, alcohol is one of the strongest predictors of wine quality, with a positive correlation. This means wines with higher alcohol content tend to receive higher quality ratings. The scatter plot demonstrates this positive relationship, though there is considerable variation, indicating that while alcohol is a significant factor, other features also play important roles in determining wine quality.

### 13. Create X, y for independent variables (all except `quality`) and target variable `quality`, perform k-fold k=3 split on them and calculate and print out the length of training set and testing set for each fold

```python
from sklearn.model_selection import KFold

X = features
y = quality

kf = KFold(n_splits=3, shuffle=True, random_state=42)

fold_num = 1
for train_index, test_index in kf.split(X):
    train_size = len(train_index)
    test_size = len(test_index)
    print(f"Fold {fold_num}:")
    print(f"  Training set size: {train_size}")
    print(f"  Testing set size: {test_size}")
    fold_num += 1
```

### 15. Create a sample dataset with dates for a whole week. Extract day of the week with `dayofweek` function (0=Monday, 6=Sunday). Additionally flag weekends by creating another new `is_weekend` feature.

```python
dates = pd.date_range(start='2026-01-01', periods=7, freq='D')
sample_data = pd.DataFrame({
    'date': dates,
    'value': np.random.rand(7) * 100
})

sample_data['dayofweek'] = sample_data['date'].dt.dayofweek

sample_data['is_weekend'] = sample_data['dayofweek'].isin([5, 6]).astype(int)

day_names = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}
sample_data['day_name'] = sample_data['dayofweek'].map(day_names)
sample_data
```
